### 2.2 Markdowns

Markdowns, with the help of illustrations whenever possible, should be written in such a way that a learner should not feel the need to look for information elsewhere. Everything required should be mentioned within the assignment.
 
Provide hyperlinks to Documentations as well so that learners could explore on their own. When providing hyperlinks, double check to make sure they don't run into an error.
 
Also, make sure the provided documentation link is relevant. I once provided a Documentation link which was for the programming language GO, whereas our assignments are made using Python.
